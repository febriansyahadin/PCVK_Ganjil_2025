# 4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16.

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
#Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2
#perhatikan koordinat titik pt1 dan pt 2
cv.rectangle(black_img, pt1=(384,1), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

```python
#Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 yang tertulis pada kode program.
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

```python
#Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang tertulis pada kode program.
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

```python
#Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

```python
#Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah ditentukan.
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500),fontFace=font, fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16 dan int32?

int16: Tipe data ini menggunakan 16 bit untuk merepresentasikan setiap nilai piksel. Rentang nilai yang dapat disimpan adalah dari -32768 hingga 32767.

int32: Tipe data ini menggunakan 32 bit untuk merepresentasikan setiap nilai piksel. Rentang nilai yang dapat disimpan adalah dari -2147483648 hingga 2147483647.

Dalam kasus pembuatan black image awal, int16 sudah cukup karena nilai 0 berada dalam rentangnya. Namun, untuk operasi pemrosesan gambar selanjutnya yang mungkin melibatkan nilai di luar 0-255, int32 atau tipe data floating-point akan lebih sesuai.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Buat canvas hitam (500x500, RGB)
img = np.zeros((500, 500, 3), dtype=np.uint8)

# Buat 5 kotak random
for i in range(5):
    # Posisi random (x1, y1, x2, y2)
    x1, y1 = np.random.randint(0, 400, size=2)
    x2, y2 = x1 + np.random.randint(30, 100), y1 + np.random.randint(30, 100)

    # Warna random (BGR format di OpenCV)
    color = tuple(np.random.randint(0, 256, size=3).tolist())

    # Gambar kotak
    cv2.rectangle(img, (x1, y1), (x2, y2), color, -1)  # -1 = isi penuh

# Tampilkan dengan matplotlib (konversi BGRRGB)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("5 Kotak Random")
plt.axis("off")
plt.show()
```

8. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB

```python
import cv2
import matplotlib.pyplot as plt

# Baca citra (contoh lena.png)
img = cv2.imread("/content/drive/MyDrive/PCVK/Image/Lenna.png")

# Pisahkan channel
B, G, R = cv2.split(img)

# Ambil ukuran gambar
h, w = B.shape

# Tambahkan rectangle di sudut kiri bawah
# (0, h-100) = kiri bawah, (100, h) = kanan bawah (rectangle 100x100 pixel)
cv2.rectangle(B, (0, h-100), (100, h), (255, 255, 255), 3)

plt.imshow(B, cmap="gray")
plt.title("Channel B dengan Rectangle di Sudut Kiri Bawah")
plt.axis("off")
plt.show()
```

9. Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran
font, dan warna font yang sesuai keinginan anda.

```python
import cv2
import matplotlib.pyplot as plt

# Path file citra
path = "/content/drive/MyDrive/PCVK/Image/Lenna.png"

# Baca gambar
image = cv2.imread(path)

# Ambil channel Biru (B)
blue_channel = image[:, :, 0]
height, width = blue_channel.shape

# Copy channel biru untuk digambar rectangle
blue_copy = blue_channel.copy()
rect_start = (0, height-100)       # sudut kiri bawah (y mendekati tinggi gambar)
rect_end = (600, height)           # panjang rectangle ke kanan
cv2.rectangle(blue_copy, rect_start, rect_end, color=0, thickness=-1)  # filled rectangle

# Ubah channel biru (grayscale) jadi 3 channel lagi
blue_rgb = cv2.cvtColor(blue_copy, cv2.COLOR_GRAY2BGR)

# Tambahkan teks nama file di atas rectangle
cv2.putText(blue_rgb,
            "Lenna.png",              # nama file
            (20, height-50),          # posisi teks
            cv2.FONT_HERSHEY_SIMPLEX, # jenis font
            2,                        # ukuran font
            (255, 255, 255),          # warna font (putih)
            5,                        # ketebalan
            cv2.LINE_AA)

# Tampilkan hasil
plt.figure(figsize=(8,6))
plt.imshow(blue_rgb)
plt.title("Nomor 9 - Channel B dengan Nama File")
plt.axis('off')
plt.show()
```
